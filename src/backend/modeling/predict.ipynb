```python
X = pd.read_csv('backend/preprocess/preprocess.csv')
week = int(input('What week is it?\t'))
X = X[(X['week'] == week) & (X['season'] == 2022)]
X.set_index(keys=['date', 'home', 'away', 'week', 'season'], inplace=True)
X.drop(['y'], axis=1, inplace=True)
X
```

```python
model = pickle.load(open('modeling/models/nn.pkl', 'rb'))
```

```python
y_prob = model.predict_proba(X)
y = pd.DataFrame(y_prob, index=X.index)
y.reset_index(['home', 'away', 'date'], drop=False, inplace=True)
y['pred'] = np.where(y[1] > y[0], y['home'], y['away'])
y.set_index(['home', 'away'], append=True, drop=True, inplace=True)
```

```python
y
```

```python
odds = pd.read_csv('backend/data/odds/odds.csv')
odds = odds[(odds['season'] == 2022) & (odds['week'] == week)]
odds['home'] = odds['home'].apply(lambda x: x.strip())
odds['away'] = odds['away'].apply(lambda x: x.strip())
odds.set_index(['week', 'season', 'home', 'away'], inplace=True)
odds = odds[['ml_h', 'ml_a']]
odds
```

```python
def risk_management(diff):
    if 0 <= diff <= .25:
        if 0 <= diff < .05:
            return 1
        elif .05 <= diff < .10:
            return 2
        elif .10 <= diff < .15:
            return 3
        elif .15 <= diff < .20:
            return 4
        elif .20 <= diff <= .25:
            return 5
    else:
        return None
```

```python
df = pd.merge(y, odds, left_index=True, right_index=True)
df['ml_perc_h'] = np.where(df['ml_h'] < 0, abs(df['ml_h']) / (abs(df['ml_h']) + 100), 100 / (abs(df['ml_h']) + 100))
df['ml_perc_a'] = np.where(df['ml_a'] < 0, abs(df['ml_a']) / (abs(df['ml_a']) + 100), 100 / (abs(df['ml_a']) + 100))
df['diff'] = np.where(df[1] > df[0], df[1] - df['ml_perc_h'], df[0] - df['ml_perc_a'])
df['ml'] = np.where(df[1] > df[0], df['ml_h'], df['ml_a'])
df['unit'] = df['diff'].apply(lambda x: risk_management(x))
```

```python
teams = {
    'Chicago Bears' : 'CHI', 'Atlanta Falcons': 'ATL', 'Cleveland Browns': 'CLE',
    'Green Bay Packers': 'GB', 'Indianapolis Colts': 'IND', 'Kansas City Chiefs': 'KC',
    'Los Angeles Rams': 'LAR', 'Miami Dolphins': 'MIA', 'New Orleans Saints': 'NO',
    'New York Giants': 'NYG', 'Philadelphia Eagles': 'PHI', 'Pittsburgh Steelers': 'PIT',
    'Los Angeles Chargers': 'LAC', 'Washington Commanders': 'WAS', 'San Francisco 49ers': 'SF', 
    'New England Patriots': 'NE', 'New York Jets': 'NYJ', 'Jacksonville Jaguars': 'JAC', 
    'Buffalo Bills': 'BUF', 'Carolina Panthers': 'CAR', 'Minnesota Vikings': 'MIN', 
    'Cincinnati Bengals': 'CIN', 'Baltimore Ravens': 'BAL', 'Dallas Cowboys': 'DAL',
    'Tampa Bay Buccaneers': 'TB', 'Denver Broncos': 'DEN', 'Las Vegas Raiders': 'LV',
    'Tennessee Titans': 'TEN', 'Detroit Lions': 'DET', 'Houston Texans': 'HOU', 'Seattle Seahawks': 'SEA',
    'Arizona Cardinals': 'ARI'
}
df = df.reset_index()
df['home'] = df['home'].apply(lambda x: teams[x])
df['away'] = df['away'].apply(lambda x: teams[x])
df['pred'] = df['pred'].apply(lambda x: teams[x])
for index, row in df.iterrows():
    print(f"{row['away'].split(' ')[0]} @ {row['home'].split(' ')[0]} {row['ml']}")
    print(f"\t{row['pred'].split(' ')[0]}")
```
